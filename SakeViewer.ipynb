```
import pandas as pd
import urllib.request
import io

#URLはCSVがUTF８で取得できれば大丈夫です。
#Google SpeadSheetであれば以下で調べるとCSVとして取得する方法がわかります。
#https://www.google.com/search?q=google+speadsheet+csv+export+url
url = "http://www17416ui.sakura.ne.jp/c4sake/getcsv.php"

#csvを読み込む関数
def read_csv(url):
    res = urllib.request.urlopen(url)
    res=res.read().decode('utf8')
    df = pd.read_csv(io.StringIO( res) )
    return df

df = read_csv(url)
```

```
import plotly.express as px

# 項目はCSVによる
fig = px.scatter_3d(df, x='アルコール度数', range_x=[12,19],
                    y='日本酒度', range_y=[-10, 10],
                    z='酸度', range_z=[0.5, 2.5], 
                    color='精米歩合', hover_name='名前')
fig.show()
```
